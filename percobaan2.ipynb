```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile
```

```python
sampling_freq, signal = wavfile.read('Aril_F55122011.wav')
```

```python
print ('\nSignal shape:', signal.shape)
print ('Datatype:', signal.dtype)
print ('Signal duration:', round(signal.shape[0] / float (sampling_freq), 2), 'second')
```

```python
signal = signal / np.power(2, 15)
```

```python
signal = signal[:1000]
```

```python
time_axis = 1000 * np.arange(0, len(signal), 1) / float(sampling_freq)
```

```python
plt.plot(time_axis, signal, color='khaki')
plt.xlabel ('Time (milliseconds)')
plt.ylabel ('Amplitude')
plt.title ('input audio signal')
plt.show()
```
